## Dictionary
Dictionaries can store multiple values like lists. In difference to lists, the values are not accessed by their position, but by name. Dictionaries can be compared to a telephone book, to the index of a book or to an inventory. For every item there are two variables, the key and the value. The key is in the telephone book example the name of the user, the value is the telephone number.

````python
phonebook = {}
phonebook['Oliver Denzel']='0000'
phonebook['BMP']='+49 7720 307-4628'

phonebook
````



If we read in the data from last time into a dictionary, we can do this as follows:

````python
data = {}
with open('de-nrw-dortmund-versorgung_mit_energie_und_wasser_seit_1994.csv', 'r', encoding='latin1') as csvfile:
    rows = csv.reader(csvfile, delimiter=';', quotechar='"')
    #read in the header line
    headers = next(rows, None)
    #generate a list of values for every heading
    for header in headers:
        data[header]=[]
    for row in rows:
        #for every entry in headers and row (first header in h, first row-entry in v and so on)
        for h,v in zip(headers, row):
            #if v is a number, convert the value
            if (v.isdigit):
                data[h].append(int(v))
            elif (v.isdecimal):
                data[h].append(float(v))
            else:  
                data[h].append(v)
data
````

## Define your own simple class
A class can store different values (=> attributes) and methods for each member (=> object) of the class.

In our example it means, that every dog has a name that is unique. So a Terrier is an object of the class dog.

It is also possible to store information on the class itself. In our example, this is an information which is common for all dogs.

````python
class dog:
    sound = 'bark'
    def __init__(self, name):
        self.name = name
    def __repr__(self):
        return "Dog: " + self.name + " which " + self.sound + "s"
````

## Use your own class
Using the class above, you can create a few objects of the class dog.

````python
terrier = dog("Terrier")
dachshund = dog("Dachshund")
lassie = dog("Collie")
print(lassie, dachshund, terrier)
terrier.sound = 'whistle'
print(lassie, dachshund, terrier)
dog.sound='call'
print(lassie, dachshund, terrier)
````
